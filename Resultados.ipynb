# Trabalho implementação operadores morfologicos

## imports

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import morphology
from morph.binary_morph_matrix import BinaryMorphMatrix
from morph.grayscale_morph_matrix import GrayscaleMorphMatrix
from morph.kernel_elements.kernel_element import KernelElement
```

# Funções

```python
def show_images_grid(w, h,*images, titles = [], cmap="gray", **fig_kwargs):
  fig = plt.figure(**fig_kwargs)
  cont = 1
  for i,image in enumerate(images):
    subplot = fig.add_subplot(w, h, cont)
    subplot.imshow(image, cmap=cmap)
    subplot.set_xticks([])
    subplot.set_yticks([])
    if len(titles) != 0:
      subplot.set_title(titles[i])
    cont+=1
```

# Comparações entre a implementação do trabalho é a do scipy

```python
base_img = BinaryMorphMatrix.from_image("test.png").data
base_kernel_element = np.ones((5, 5))
kernel_element = KernelElement(base_kernel_element, [3, 3])
```

```python
# imagem original
plt.imshow(base_img, cmap="gray")
```

### Operador de erosão

```python
img_scipy = morphology.binary_erosion(base_img, base_kernel_element)
img_implement = BinaryMorphMatrix(base_img).erosion(kernel_element).data

show_images_grid(1, 3, base_img, img_scipy, img_implement, titles=["Imagem original","Erosão scipy", "Erosão implementada"], figsize = (12, 8))
```

### Operador de dilatação

```python
img_scipy = morphology.binary_dilation(base_img, base_kernel_element)
img_implement = BinaryMorphMatrix(base_img).dilation(kernel_element).data

show_images_grid(1, 3, base_img, img_scipy, img_implement, titles=["Imagem original","Dilatação scipy", "Dilatação implementada"], figsize = (12, 8))
```

### Operador de abertura

```python
base_img = BinaryMorphMatrix.from_image("test_opening.png").data
img_scipy = morphology.binary_opening(base_img, base_kernel_element)
img_implement = BinaryMorphMatrix(base_img).opening(kernel_element).data

show_images_grid(1, 3, base_img, img_scipy, img_implement, titles=["Imagem original","Abertura scipy", "Abertura implementada"], figsize = (12, 8))
```

### Operador de fechamento

```python
base_img = BinaryMorphMatrix.from_image("test_closing.png").data
img_scipy = morphology.binary_closing(base_img, base_kernel_element)
img_implement = BinaryMorphMatrix(base_img).closing(kernel_element).data

show_images_grid(1, 3, base_img, img_scipy, img_implement, titles=["Imagem original","Fechamento scipy", "Fechamento implementada"], figsize = (12, 8))
```

# Comparação entra a implmentação com imagens em grayscale

### Operador de abertura

```python
img_scipy = morphology.grey_opening(base_img, size=(5, 5))
img_implement = GrayscaleMorphMatrix(base_img).opening(kernel_element).data

show_images_grid(1, 3, base_img, img_scipy, img_implement, titles=["Imagem original","Abertura scipy", "Abertura implementada"], figsize = (12, 8))
```
